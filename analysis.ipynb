```python
import pandas as pd 
import matplotlib.pyplot as plt
accident_df = pd.read_csv("110accident.csv")
accident_df[0:25]
```

```python
accident_df.shape
```

```python
month_list = [0]*12

for i in accident_df["發生日期"] :
    """
    if len(str(i)) != 8 :
        print("error")
    if str(i)[0] != '2':
        print(str(i)[0])
        print("error2")
    """
    month_str = str(i)[4:6]
    month_list[int(month_str) - 1] += 1
    
        
```

```python
%matplotlib inline
print(month_list)
x_month = [*range(1,13)]
print(x_month)
plt.xlabel("month(1-12)")
plt.ylabel("accident counts")
plt.title("2021 accidents per month")
plt.bar(x_month , height=month_list , color = ['dodgerblue' , 'dodgerblue' , 'dodgerblue' , 'steelblue', 'steelblue', 'steelblue' , 'cadetblue' , 'cadetblue', 'cadetblue' , 'mediumspringgreen' , 'mediumspringgreen', 'mediumspringgreen'] 
        , width = 0.7)
plt.savefig('2021_accident_per_manth.png') 
plt.show()
```

```python
time_col = accident_df["發生時間"]
list_t = []
for t in time_col :
    list_t.append(len(str(t)))
    
print(max(list_t) , min(list_t))
```

```python
new_hour_list = []
for t in time_col :
    space = 0
    new_time_str = ""
    str_t = str(t)
    if len(str(t)) != 6 :
         space = 6 - len(str_t)
    if space == 0 :
        new_time_str = str_t
    if space == 1 : 
        new_time_str = "0" + str_t
    elif space == 2 :
        new_time_str = "00" + str_t
    elif space == 3 :
        new_time_str = "000" + str_t
    elif space == 4 :
        new_time_str = "0000" + str_t
    elif space == 5 :
        new_time_str = "000000" + str_t
    elif space == 6 :
        print("error")
    elif space == 7 :
        print("error")
    #print(str(t) , ',', new_time_str)
    new_hour_list.append(new_time_str[0:2])
```

```python
new_hour_list
newhour_count = [0]*24
for t in new_hour_list :
    newhour_count[int(t)] += 1
```

```python
newhour_count
```

```python
plt.xlabel("time(00:00-23:00)")
plt.ylabel("accident counts")
plt.title("2021 accidents per hour")
plt.bar([*range(0,24)] , height=newhour_count , color = ['dodgerblue', 'steelblue' , 'cadetblue' , 'mediumspringgreen' ,
                                                         'dodgerblue', 'steelblue' , 'cadetblue' , 'red' ,
                                                         'dodgerblue', 'steelblue' , 'cadetblue' , 'mediumspringgreen' ,
                                                         'dodgerblue', 'steelblue' , 'cadetblue' , 'mediumspringgreen' ,
                                                         'dodgerblue', 'red' , 'cadetblue' , 'mediumspringgreen' ,
                                                         'dodgerblue', 'steelblue' , 'cadetblue' , 'mediumspringgreen' ,])
plt.savefig('2021_accident_per_hour.png') 
plt.show()
```
